# Importar Librerías a utilizar

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

# Split Test-Train-Validation

```python
df = pd.read_csv('/home/bryan/Escritorio/Paper/Prueba1/HB.csv')
```

```python
df.head()
```

```python
X_train, x_tv = train_test_split(df, test_size=0.30, shuffle=False)
```

```python
x_test, x_val = train_test_split(x_tv.iloc[:,1:], test_size=0.10, shuffle=False)
```

```python
X_train.iloc[:,1:].to_csv('/home/bryan/Escritorio/Paper/Entrenamiento/Train.csv')
x_test.to_csv('/home/bryan/Escritorio/Paper/Entrenamiento/Test.csv')
x_val.to_csv('/home/bryan/Escritorio/Paper/Entrenamiento/Val.csv')
```

# Reescalar

```python
train = pd.read_csv('/home/bryan/Escritorio/Paper/Entrenamiento/Train.csv').iloc[:,1:]
```

```python
train.head()
```

```python
scaler = MinMaxScaler()
```

```python
scaled_train = scaler.fit_transform(train)
```

```python
test = pd.read_csv('/home/bryan/Escritorio/Paper/Entrenamiento/Test.csv').iloc[:,1:]
```

```python
scaled_test = scaler.fit_transform(test)
```

```python
val = pd.read_csv('/home/bryan/Escritorio/Paper/Entrenamiento/Val.csv').iloc[:,1:]
```

```python
scaled_val = scaler.fit_transform(val)
```

### Guardar lo reescalado

```python
def cols_exp():
    index_ex = []
    for sensor in range(1,9):
            index_ex.append('-X'+str(sensor)+'-')
            index_ex.append('-Y'+str(sensor)+'-')
            index_ex.append('-Z'+str(sensor)+'-')
    return index_ex
```

```python
df_t = pd.DataFrame(scaled_train, columns = cols_exp()).to_csv('/home/bryan/Escritorio/Paper/Reescalado/Train.csv')
df_ts = pd.DataFrame(scaled_test, columns = cols_exp()).to_csv('/home/bryan/Escritorio/Paper/Reescalado/Test.csv')
df_v = pd.DataFrame(scaled_val, columns = cols_exp()).to_csv('/home/bryan/Escritorio/Paper/Reescalado/Val.csv')
```
