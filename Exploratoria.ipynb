```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Recorrer directorio

```python
#Directorio a recorrer
path_mat = '/home/bryan/Escritorio/Paper/DataCSV/'

def Ficheros(path, formato):
    contenido = os.listdir(path)

    #Acumulador del nombre de las   imagenes a utilizar
    archivos = []

    #Se recorre el path para obtener los nombre y se los aggrega al acumulador, en este caso la lista
    for fichero in contenido:
        if os.path.isfile(os.path.join(path, fichero)) and fichero.endswith(formato):
            archivos.append(fichero)
            
    return archivos


#Verificar si obtiene los archivos
print(len(Ficheros(path_mat, '.csv')))
```

```python
archivos = Ficheros(path_mat, '.csv')
prueba_1 = []
prueba_4 = []
for archivo in archivos:
    if archivo.startswith('1_'):
        prueba_1.append(archivo)
    
    elif archivo.startswith('4_'):
        prueba_4.append(archivo)

prueba_1.sort()
prueba_4.sort()
```

```python
print(len(prueba_1), len(prueba_4))
```

```python
def index_exp(archivos, sensor):
    index_ex = []
    for archivo in archivos:
        index_ex.append('X'+str(sensor)+'-'+ archivo[:-4])
        index_ex.append('Y'+str(sensor)+'-'+ archivo[:-4])
        index_ex.append('Z'+str(sensor)+'-'+ archivo[:-4])
    return index_ex
```

```python
def dataframe_chipeado(lista, sensor):
    media = []
    mediana = []
    moda = []
    std = []
    minimo = []
    maximo = []

    for csv in range(len(lista)):
        df_t = pd.read_csv(path_mat + lista[csv])
        a1 = df_t.iloc[:,1:4]

        a = a1.describe().transpose()
        me = a['mean'].values.tolist()
        med = a1.median().values.tolist()
        mod = a1.mode().values.tolist()[0]

        st = a['std'].values.tolist()
        mi = a['min'].values.tolist()
        ma = a['max'].values.tolist()

        for j in range(3):
            media.append(me[j])
            mediana.append(med[j])
            moda.append(mod[j])

            std.append(st[j])
            minimo.append(mi[j])
            maximo.append(ma[j])
    
    indices = index_exp(lista, sensor)

    df_HB = {}
    df_HB= pd.DataFrame(df_HB, columns=['Prueba','Media', 'Mediana', 'Moda','STD',  'Min', 'Max'])
    df_HB['Media'], df_HB['Mediana'], df_HB['Moda'], df_HB['STD'], df_HB['Min'], df_HB['Max'], df_HB['Prueba'] = media, mediana, moda, std, minimo, maximo, indices
    
    return df_HB
```
